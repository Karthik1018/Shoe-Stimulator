```python
from PIL import Image, ImageDraw
count =0
import time
while True:
    f=open('s1.txt','r')
    s1=f.read()
    f.close()
    f=open('s2.txt','r')
    s2=f.read()
    f.close()
    f=open('s3.txt','r')
    s3=f.read()
    f.close()
    f=open('s4.txt','r')
    s4=f.read()
    f.close()
    f=open('s5.txt','r')
    s5=f.read()
    f.close()
    
    s1,s2,s3,s4,s5=int(s1),int(s2),int(s3),int(s4),int(s5)
    print(s1,s2,s3,s4,s5)
    
    # Open the image file
    image = Image.open("foot.png")

    # Create a new ImageDraw object
    draw = ImageDraw.Draw(image)

    # Define the box coordinates (left, top, right, bottom)

    box1 = (150, 1600, 500, 2000)
    
    if s1<=50:
        circle_color1 = (0, 255, 0, 128)
    elif s1<=100:
        circle_color1 = (255, 234, 0, 128)
    else:
        circle_color1 = (255, 0, 0, 128)
   
        
    if s2<=50:
        circle_color2 = (0, 255, 0, 128)
    elif s2<=100:
        circle_color2 = (255, 234, 0, 128)
    else:
        circle_color2 = (255, 0, 0, 128)
    
    if s3<=100:
        circle_color3 = (0, 255, 0, 128)
    elif s3<=200:
        circle_color3 = (255, 234, 0, 128)
    else:
        circle_color3 = (255, 0, 0, 128)
    
    if s4<=200:
        circle_color4 = (0, 255, 0, 128)
    elif s4<=300:
        circle_color4 = (255, 234, 0, 128)
    else:
        circle_color4 = (255, 0, 0, 128)
    
    if s5<=30:
        box_color1 = (0, 255, 0, 128)
    elif s5<=50:
        box_color1 = (255, 234, 0, 128)
    else:
        box_color1 = (255, 0, 0, 128)

    center1 = (250, 200)
    center2 = (250, 700)
    center3 = (590, 850)
    center4 = (590, 1300)
    radius = 100

    

    # Draw the circle on the image
    draw.ellipse((center1[0]-radius, center1[1]-radius, center1[0]+radius, center1[1]+radius), fill=circle_color1)
    draw.ellipse((center2[0]-radius, center2[1]-radius, center2[0]+radius, center2[1]+radius), fill=circle_color2)
    draw.ellipse((center3[0]-radius, center3[1]-radius, center3[0]+radius, center3[1]+radius), fill=circle_color3)
    draw.ellipse((center4[0]-radius, center4[1]-radius, center4[0]+radius, center4[1]+radius), fill=circle_color4)


    # Draw the box on the image using a red color and a thickness of 3 pixels
    #draw.rectangle(box, outline=(255, 0, 0), width=3,fill=box_color)
    draw.rectangle(box1, outline=(0, 255, 0), width=3,fill=box_color1)
    
    # Show the image
    #image.show()
    image.save(str(count)+".png")
    f=open('count.txt','w')
    f.write(str(count))
    f.close()
    count =count+1
    print('count',count)
    f=open('D:/ml_code/input.txt','w')
    f.write(str(s1)+'\n'+str(s2)+'\n'+str(s3)+'\n'+str(s4)+'\n'+str(s5))
    f.close()
    time.sleep(10)
```
